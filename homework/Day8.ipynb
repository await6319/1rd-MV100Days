```python
import cv2
import numpy as np

img_path = "D:/lena.png"
#img_path = "image/01.jpg"
img = cv2.imread(img_path)

img_blur = img.copy()
#################################模糊#################################
# 重複多次 Gaussian 模糊的操作來加深模糊的程度
img_blur = cv2.GaussianBlur(img_blur, (5, 5), 0)
img_blur = cv2.GaussianBlur(img_blur, (5, 5), 0)
img_blur = cv2.GaussianBlur(img_blur, (5, 5), 0)

#################################邊緣檢測#################################
img_grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 對 x 方向做 Sobel 邊緣檢測
img_sobel_x = cv2.Sobel(img_grey, cv2.CV_16S, dx=1, dy=0, ksize=3)
img_sobel_x = cv2.convertScaleAbs(img_sobel_x)

# 對 y 方向做 Sobel 邊緣檢測
img_sobel_y = cv2.Sobel(img_grey, cv2.CV_16S, dx=0, dy=1, ksize=3)
img_sobel_y = cv2.convertScaleAbs(img_sobel_y)

# x, y 方向的邊緣檢測後的圖各以一半的全重進行合成
img_sobel_combine = cv2.addWeighted(img_sobel_x, 0.5, img_sobel_y, 0.5, 0)

#加入文字標示
cv2.putText(img_sobel_x, '(X)', (30, 40), 0, 1, (255, 255, 255), 2)
cv2.putText(img_sobel_y, '(Y)', (30, 40), 0, 1, (255, 255, 255), 2)
cv2.putText(img_sobel_combine, '(X+Y)', (30, 40), 0, 1, (255, 255, 255), 2)

"""
作業
實作模糊與邊緣檢測

透過 Gaussian Filter 實作模糊操作
透過 Sobel Filter 實作邊緣檢測

"""
#################################作業一＿uint8 Sobel #################################
# 轉為灰階圖片
img_grey1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# 對 x 方向以包含負數的資料格式 (cv2.CV_16S) 進行 Sobel 邊緣檢測
img_sobel_x1 =cv2.Sobel(img_grey1, cv2.CV_16S, dx=1, dy=0, ksize=3)

# 對 x 方向依照比例縮放到所有數值都是非負整數
img_sobel_x1 = cv2.convertScaleAbs(img_sobel_x1)

# 對 x 方向直接以非負整數的資料格式 (uint8) 進行 Sobel 邊緣檢測
img_sobel_x_uint8 = cv2.Sobel(img_grey1,cv2.CV_8U, dx=1, dy=0, ksize=3)

#加入文字標示
cv2.putText(img_sobel_x1, '(X_CV_16S)', (30, 40), 0, 1, (255, 255, 255), 2)
cv2.putText(img_sobel_x_uint8, '(X_uint8)', (30, 40), 0, 1, (255, 255, 255), 2)

#################################作業二＿一次與兩次計算偏微分#################################
# 轉為灰階圖片
img_grey2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 求一次導數取得邊緣檢測結果
img_sobel_x2 = cv2.Sobel(img_grey1, cv2.CV_16S, dx=1, dy=0, ksize=3)
img_sobel_x2 = cv2.convertScaleAbs(img_sobel_x2)
# 求二次導數取得邊緣檢測結果
img_sobel_xx2 = cv2.Sobel(img_grey1, cv2.CV_16S, dx=2, dy=0, ksize=3)
img_sobel_xx2 = cv2.convertScaleAbs(img_sobel_xx2)

#　組合 + 顯示圖片
img_show = np.hstack((img, img_blur))
img_sobel = np.hstack((img_sobel_x, img_sobel_y, img_sobel_combine))
img_sobel_uint8 = np.hstack((img_grey, img_sobel_x1, img_sobel_x_uint8))
img_sobel_xx=np.hstack((img_grey,img_sobel_x2,img_sobel_xx2))
while True:
    cv2.imshow('Gaussian Blur', img_show)
    cv2.imshow('Edge Detection', img_sobel)
    cv2.imshow('Edge Detection_uint8', img_sobel_uint8)
    cv2.imshow('Edge Detection_xx',img_sobel_xx)
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
```
